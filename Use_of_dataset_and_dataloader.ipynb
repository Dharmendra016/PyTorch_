```python
import torch
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from torch.utils.data import Dataset, DataLoader
```

```python
df.drop(columns=['id' , 'Unnamed: 32'], inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
#scalling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) # use transform not fit_transform as
#we use the same scaler We should use the same scaling parameters (mean & standard deviation) from the training data.
```

```python
#label encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train= torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

```python
class CustomDataset(Dataset):
  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, idx):
    return self.features[idx], self.labels[idx]
```

```python
train_dataset = CustomDataset(X_train, y_train)
test_dataset = CustomDataset(X_test, y_test)
```

```python
train_dataset[0]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True , drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, drop_last=True)
```

```python
#defining the model
import torch.nn as nn
class MySimpleNN(nn.Module):

  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features , 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self , x):
    x = self.linear(x)
    x = self.sigmoid(x)
    return x
```

```python
#loss function
loss_function = nn.BCELoss()
```

```python
learning_rate = 0.1
epochs = 50
```

```python
#training pipeline
#create model
model = MySimpleNN(X_train.shape[1])

#define optimizer
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python

#define loop for epochs
for epoch in range(epochs):

  for batch_features , batch_labels in train_loader:

      #forward pass
      y_pred = model(batch_features)

      #loss calculate
      loss = loss_function(y_pred , batch_labels.view(-1,1))  #.view(-1,1)  shape will be (row_of_y_train , 1) in this case (455,1)

      #clear grad
      optimizer.zero_grad()

      #backward pass
      loss.backward()

      #parameters update
      optimizer.step()

  #print loss in each epoch
  print(f'Epoch: {epoch + 1} Loss: {loss.item()}')
```

```python
#when model is initialized then model.parameters will register all the weights and bias in the form of interations like as below:
for i in model.parameters():
  print(i)
```

```python
#evaluation
with torch.no_grad():
  for batch_features , batch_labels in test_loader:
    y_pred = model.forward(batch_features)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == batch_labels).float().mean()
    print(f'Accuracy: {accuracy}')
```
