```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
#scalling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test) # use transform not fit_transform as
#we use the same scaler We should use the same scaling parameters (mean & standard deviation) from the training data.
```

```python
#label encoding
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
#Numpy arrays to PyTorch tensors
X_train = torch.from_numpy(X_train)
X_test = torch.from_numpy(X_test)
y_train = torch.from_numpy(y_test)
y_test = torch.from_numpy(y_test)
```

```python
X_train
```

```python
#defining the model

class MySimpleNN():

  def __init__(self, X):

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64,requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64 , requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X , self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self , y_pred , y):
    #clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred , epsilon , 1-epsilon)

    loss = -y * torch.log(y_pred) - (1-y) * torch.log(1-y_pred)
    return loss.mean()

  def update_parameters(self,learning_rate):
    with torch.no_grad():
      self.weights -= learning_rate * self.weights.grad
      self.bias -= learning_rate * self.bias.grad
```

```python
learning_rate = 0.1
epochs = 50
```

```python
#training pipeline
#create model
model = MySimpleNN(X_train)

#define loop for epochs
for epoch in range(epochs):
  #forward pass
  y_pred = model.forward(X_train)

  #loss calculate
  loss = model.loss_function(y_pred , y_train)

  #backward pass
  loss.backward()

  #parameters update
  model.update_parameters(learning_rate)

  #reset gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  #print loss in each epoch
  print(f'Epoch: {epoch + 1} Loss: {loss.item()}')
```

```python
#evaluation
with torch.no_grad():
  y_pred = model.forward(X_test)
  y_pred = (y_pred > 0.5).float()
  accuracy = (y_pred == y_test).float().mean()
  print(f'Accuracy: {accuracy}')
```
